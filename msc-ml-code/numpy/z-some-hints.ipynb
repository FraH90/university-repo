## Alayws specify the type of vector (colum or row)


```python
import numpy as np

vect = np.array([1, 2, 3])
vect.shape

vect_col = vect.reshape(-1, 1)
vect_row = vect.reshape(1, -1)

print(f'Column vector, shape {vect_col.shape}:\n {vect_col}\n')
print(f'Row vector, shape {vect_row.shape}:\n {vect_row}\n')
print(f'Transpose of the column vector:\n {vect_col.T}\n')
print(f'Transpose of the row vector:\n {vect_row.T}\n')

```
